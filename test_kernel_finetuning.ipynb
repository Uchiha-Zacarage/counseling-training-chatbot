```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from PIL import Image
from IPython.display import display
from google.colab.patches import cv2_imshow
import os

from google.colab import drive
drive.mount('/content/drive')
```

```python
def create_kernel(ksize=5, sigma=3, theta=np.pi/4, lamda=np.pi/4, gamma=0.5, phi=0):
  # Create the Gabor filter
  if ksize%2==0: ksize += 1
  kernel = cv2.getGaborKernel((ksize, ksize), sigma, theta, lamda, gamma, phi, ktype=cv2.CV_32F)
  return kernel
```

```python
# Opening the folder for the images and getting all the image names
img_folder = "/content/drive/MyDrive/Neurosynth/test-images/"
images = os.listdir(img_folder)

# Getting the kernel
kernel = create_kernel(8, 5, np.pi/4, np.pi/4, 1, 0.5)

for img in images:
  # Getting the image path
  img_path = os.path.join(img_folder, img)
  tmp = cv2.imread(img_path)

  # Grayscaling the image
  gray_img = cv2.cvtColor(tmp, cv2.COLOR_BGR2GRAY)
  cv2_imshow(gray_img)

  # Filtering the image with the Gabor Filter kernel
  fil_img = cv2.filter2D(gray_img, cv2.CV_8UC3, kernel)
  cv2_imshow(fil_img)
  cv2.waitKey(0)

cv2.destroyAllWindows()
```
